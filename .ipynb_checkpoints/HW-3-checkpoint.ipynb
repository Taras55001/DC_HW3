```python
#напишіть функцію гіпотези лінійної регресії у векторному вигляді;
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from random import uniform, sample
from numpy.linalg import inv

def hypothesis(theta, X):
        return X @ theta
```

```python
#створіть функцію для обчислення функції втрат у векторному вигляді;
def compute_cost(theta, X, Y):
    m = len(Y)
    h = hypothesis(theta, X)
    error = h - Y
    return (1 / (2 * m)) * np.sum(error**2)
```

```python
#реалізуйте один крок градієнтного спуску;
def gradient_descent_step(theta, X, Y, alpha):
    m = len(Y)
    h = hypothesis(theta, X)
    error = h - Y
    gradient = (1 / m) * (X.T @ error)
    theta = theta - alpha * gradient
    return theta
```

```python
#знайдіть найкращі параметри w для датасету прогнозуючу ціну на будинок залежно від площі, кількості ванних кімнат та кількості спалень;
import pandas as pd
file_path = "./Housing.csv"
df = pd.read_csv(file_path)
sns.set_style("whitegrid")
custom_palette = sns.color_palette("deep", 6)
sns.pairplot(df, hue="bedrooms", palette=custom_palette)
plt.show()
```

```python
selected_features = ['area']
X = df[selected_features].values
y = df['price'].values
theta = np.zeros(X.shape[1])
alpha = 0.0001
num_iterations = 200
for i in range(num_iterations):
    theta = gradient_descent_step(theta, X, y, alpha)
    print("Найкращі параметри w:", theta)

#знайдіть ці ж параметри за допомогою аналітичного рішення;
theta_analytical = inv(X.T @ X) @ X.T @ y
print("Параметри w за допомогою аналітичного рішення:", theta_analytical)
```
